```python
import ADV_Grid_Class as advg 
import Turbine_Class
import Flume_Class
import pickle

flume_depth, flume_width, flume_length = 0.40, 0.77, 7.97 # The dimensions of the flume [m]
flume = Flume_Class.flume(flume_depth, flume_width, flume_length) # We create an instance of the flume class (depth, width, length)

SM_length, SM_width, SM_blade_radius, SM_axis_of_rotation, SM_tip_speed_ratio, SM_hub_height = 37, 14, 15.5, 'horizontal', 4.3, 17 # The dimensions of the Seamap turbine [cm]
Seamap = Turbine_Class.turbine(SM_length, SM_width, SM_blade_radius, SM_axis_of_rotation, SM_tip_speed_ratio, SM_hub_height) # We create a Seamap instance of the turbine class (length, width, blade_radius, axis_of_rotation, tip_speed_ratio, hub_height)

V_length, V_width, V_blade_radius, V_axis_of_rotation, V_tip_speed_ratio, V_hub_height = 70, 40, 17, 'vertical', 1, 21 # The dimensions of the Volturnus [cm]
Volturnus = Turbine_Class.turbine(V_length, V_width, V_blade_radius, V_axis_of_rotation, V_tip_speed_ratio, V_hub_height) # We create a Volturnus instance of the turbine class (length, width, blade_radius, axis_of_rotation, tip_speed_ratio, hub_height)

# Now we can create the grid
# paths = [r'ADV_Data_Seamap_1234_28Hz_Complete', r'ADV_Data_Volturnus_1234_28Hz_Complete'] # The path to the grid
# grid_names = ['Seamap 1234, 28 Hz', 'Volturnus 1234, 28 Hz'] # The name of the grid
# turbines = [Seamap, Volturnus] # The turbines used in the grid
# save_paths = ['Seamap 1234, 28 Hz_mPST_ALL_BEST.pkl', 'Volturnus 1234, 28 Hz_mPST_ALL_BEST.pkl'] # The path to save the grid

paths = [r'Seamap_HighRes_1234_No_Turbine_28Hz', r'Seamap_HighRes_23_No_Turbine_22Hz', r'Seamap_HighRes_34_No_Turbine_22Hz'] # The path to the grid
grid_names = ['No Turbine 1234, 28 Hz', 'No Turbine 23, 22 Hz', 'No Turbine 34, 22 Hz'] # The name of the grid
turbines = [Seamap, Seamap, Seamap] # The turbines used in the grid
save_paths = ['No Turbine 1234, 28 Hz_mPST.pkl', 'No Turbine 23, 22 Hz_mPST.pkl', 'No Turbine 34, 22 Hz_mPST.pkl'] # The path to save the grid

gridNT1234 = advg.ADV_Grid(paths[0], grid_names[0], turbines[0], flume)
gridNT23 = advg.ADV_Grid(paths[1], grid_names[1], turbines[1], flume)
gridNT34 = advg.ADV_Grid(paths[2], grid_names[2], turbines[2], flume)

#gridSM = advg.ADV_Grid(paths[0], grid_names[0], turbines[0], flume) # We create an instance of the grid class
#gridV = advg.ADV_Grid(paths[1], grid_names[1], turbines[1], flume)
```
